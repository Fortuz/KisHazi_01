# Task 1: Picture manipulation

```python
images=["img_01.jpg","img_02.jpg","img_03.jpg","img_04.jpg","img_05.jpg"] #This list contains the names of the 5 images
pictures = {} 
key = 0 
for image in images:          # The for loop goes over the list and imports all the images
    img = Image.open(image)   #img = Image.open('img_01.jpg')
    arr = np.array(img)       # 32x32x3 array arr = np.array(img)
    pictures[key] = arr       # The images tored in a dictionary
    key = key + 1             #The keys will be numbers 0-4
```

```python
print("Elemets in array: %d" % arr.size) # Check the imported image size and shape (debug purposes)
print("Shape of the array: %s" % str(arr.shape))
```

```python
for key in pictures:          # print out the pictures and the R-G-B matrixes
    PLT.imshow(pictures[key])
    PLT.show()
    R = pictures[key][:,:,0]
    G = pictures[key][:,:,1]
    B = pictures[key][:,:,2]
    print("R matrix of the picture")
    print(R)
    print("\n")
    print("G matrix of the picture")
    print(G)
    print("\n")
    print("B matrix of the picture")
    print(B)
    print("\n")
```

```python
RGB_list = []#empty list initialization
R_list = []
G_list = []
B_list = []
for key in pictures:
    R_flat = np.array((pictures[key][:,:,0]).flatten())       #Flatten the R component of the image into 1D array
    G_flat = np.array((pictures[key][:,:,1]).flatten())       #Flatten the G component of the image into 1D array
    B_flat = np.array((pictures[key][:,:,2]).flatten())       #Flatten the B component of the image into 1D array
    Flat = np.concatenate((R_flat, G_flat, B_flat), axis=0)   #Concatenate the 3 component
    RGB_list.append(Flat)                                     #Append to the previously empt list
    R_list.append(R_flat)
    G_list.append(G_flat)
    B_list.append(B_flat)
RGB_array = np.array(RGB_list)                                #Transform the list into a numpy array
R_array = np.array(R_list)
G_array = np.array(G_list)
B_array = np.array(B_list)
print(RGB_array)
print("\n")
print("The dimensions of the data: " + str(RGB_array.shape))  #Check the dimension 32*32*3=3072, 5 pics
print("\n")
print(R_array)
print("\n")
print("The dimensions of one chanel: " + str(R_array.shape))
```

```python
#Calculating the mean for every RGB channel
R_mean = np.mean(R_array)
G_mean = np.mean(G_array)
B_mean = np.mean(B_array)
print("R chanel mean = " + str(R_mean))
print("G chanel mean = " + str(G_mean))
print("B chanel mean = " + str(B_mean))
```

```python
#Calculating the standard deviation for every RGB channel
R_std = np.std(R_array)
G_std = np.std(G_array)
B_std = np.std(B_array)
print("R chanel std = " + str(R_std))
print("G chanel std = " + str(G_std))
print("B chanel std = " + str(B_std))
```

```python
R_norm = (R_array - R_mean)/R_std # Normalization of every chanel
G_norm = (G_array - G_mean)/G_std
B_norm = (B_array - B_mean)/B_std
print(R_norm)
print("\n")
print(R_norm.shape) # Debug: check the dimensions of the array
```

```python
#Making an array wich contains the images in one row, RGB channels are normalized and flattend
RGB_norm = np.concatenate((R_norm, G_norm, B_norm), axis=1)
print("Dimensions of RGB_norm matrix: " + str(RGB_norm.shape))
print("\n")
print("Final normalizad database is:")
print("\n")
print(RGB_norm)
```

# Task 3: Text manipulation

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup # Import some package to text manipulation
```

```python
url = "https://www.snopes.com/fact-check/facebook-ai-developed-own-language/" # URL of the desired page
html = urlopen(url) 
```

```python
soup = BeautifulSoup(html, 'lxml') 
```

```python
title = soup.title # Get the title of the page (self check for debug)
print(title) 
```

```python
for tag in soup.find_all(['script', 'style']): #Clean up the text, decompose all html connected code
    tag.decompose()
text = soup.get_text(strip=True)
print(text)                                    #plot raw text from source
```

```python
from string import ascii_lowercase     # ascii_lowercase =='abcdefghijklmnopqrstuvwxyz'
```

```python
text = text.lower() #convert the text lowercase (easier counting)
sum=0
for key in ascii_lowercase:
    print(key, text.count(key)) 
    sum=sum+text.count(key)     # checksum calculation for debug
```

```python
print("Counted characters: %d" % sum) #sum < len(text) because there are special characters
print("Length of the clear text: %d" % len(text))
```

```python
from collections import Counter #import to count characters
```

```python
letter_counts = Counter(text)
df = pd.DataFrame.from_dict(letter_counts, orient='index')  #DataFrame using Pandas
df = df.filter(items=ascii_lowercase, axis=0)               #filtering the alphabet
df.plot(kind='bar', figsize=(20,10))
PLT.show() # plot the histogram
```
